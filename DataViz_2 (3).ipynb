```python
df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/14_SeveralIndepLists.csv", sep=" ")

to_remove = "_[0123456789]|\\.|function|^id|script|var|div|null|typeof|opts|if|^r$|undefined|false|loaded|true|settimeout|eval|else|artist"

for i in range(len(df)):
    if str(df["word"][i]) in to_remove or "." in str(df["word"][i]):
        df=df.drop(index=i)

df=df.reset_index()
df
```

```python
plt.figure(figsize=(8,10))

word_counts1 = df[df['artist']=='nekfeu'].groupby('word').count()['artist'].reset_index()
word_counts = df[df['artist']=='nekfeu'].groupby('word').count()['artist'].reset_index()

word_counts = word_counts.sort_values(by='artist', ascending=False).head(10)
word_counts = word_counts.sort_values(by='artist')
word_counts
plt.figure(figsize=(8,5))

plt.hlines(y=word_counts["word"], xmin=0, xmax=word_counts["artist"])
plt.plot(word_counts["artist"],word_counts["word"], "o")
```

```python
word_counts=word_counts.reset_index()
word_counts
```

```python
text=""

for i in range(10):
    text+= ((word_counts["word"][i]+" ")*word_counts["artist"][i])

text
```

```python
from wordcloud import WordCloud

# Create the word cloud object
wordcloud = WordCloud().generate(text)

# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from matplotlib_venn import venn3

a = df[df['artist']=='nekfeu'].groupby('word').count()['artist'].reset_index()
b = df[df['artist']=='booba'].groupby('word').count()['artist'].reset_index()
c = df[df['artist']=='georges-brassens'].groupby('word').count()['artist'].reset_index()

set1 = set(a["word"])
set2 = set(b["word"])
set3 = set(c["word"])

# create venn diagram
venn = venn3([set1, set2, set3])

# display the diagram
plt.show()

print("A: Nekfeu \nB: Booba\nC: Brassens")
```

```python
sns.heatmap(df,cmap='Blues')
```

```python
a=df.columns.to_list()
ab=10*a
ab
```

```python
df_melted = pd.melt(df, value_vars=df.columns)
df_melted.rename(columns = {'variable':'source'}, inplace = True)
df_melted['target']=ab

df_melted
```

```python
for i in a:
    df_melted=df_melted.replace(i,a.index(i))
df_melted
```

```python
import plotly.graph_objects as go

# Define the data for the chart
data = dict(
    type='sankey',
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(
        color = "black",
        width = 0.5
      ),
        label = ['Africa',
 'East Asia',
 'Europe',
 'Latin America',
 'North America',
 'Oceania',
 'South Asia',
 'South East Asia',
 'Soviet Union',
 'West Asia'],
    color = ["blue", "red", "green", "black", "cyan", "pink","orange","yellow","maroon","lime"]
    ),
    link = dict(
      source = df_melted['source'],
      target = df_melted['target'],
      value = df_melted['value']
  ))

# Create the layout for the chart
layout =  dict(
    title = "Basic Sankey Diagram",
    font = dict(
      size = 10
    )
)

# Create the figure
fig = go.Figure(data=[data], layout=layout)

# Show the figure
fig.show()
```

####several obs per grp (subgrp)

```python
df=pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/10_OneNumSevCatSubgroupsSevObs.csv", sep=",")
df.dropna()

df["tip_perc"]=[100*df['tip'][i]/df['total_bill'][i] for i in range(len(df))]
df
```

```python
plt.figure(figsize=(8,6))

df["day"] = df["day"].astype("category")
df["day"].cat.reorder_categories(["Thur", "Fri", "Sat", "Sun"], inplace=True)

# Create the grouped box plot
sns.boxplot(x="day", y="tip_perc", hue="sex", data=df, dodge=True, width=0.5, fliersize=0)

plt.show()
```

```python
plt.figure(figsize=(8,6))

df["sex"] = df["sex"].astype("category")
df["sex"].cat.reorder_categories(["Male", "Female"], inplace=True)

# Create the grouped box plot
sns.boxplot(x="sex", y="tip_perc", hue="day", data=df, dodge=True, width=0.5, fliersize=0)

plt.show()
```

```python
plt.figure(figsize=(8,6))

df["day"] = df["day"].astype("category")
df["day"].cat.reorder_categories(["Thur", "Fri", "Sat", "Sun"], inplace=True)

sns.violinplot(x="day", y="tip_perc", hue="sex", data=df, dodge=True, width=0.5, fliersize=0)

plt.show()
```

```python
plt.figure(figsize=(8,6))

df["sex"] = df["sex"].astype("category")
df["sex"].cat.reorder_categories(["Male", "Female"], inplace=True)

sns.violinplot(x="sex", y="tip_perc", hue="day", data=df, dodge=True, width=0.5, fliersize=0)

plt.show()
```

```python
df["time"] = df["time"].astype("category")
df["time"].cat.reorder_categories(["Lunch", "Dinner"], inplace=True)

a = sns.FacetGrid(data=df, row="sex", col="time")
a.map(sns.histplot, "tip_perc", bins=20, kde=False, fill="#69b3a2", color="black")
```
